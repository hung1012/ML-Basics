# Pyplot API vs Object-Oriented API
* Pyplot API -> Quickly
* OO API -> Advanced

```python
# Pyplot API
x = np.linspace(0,10, 1000)
plt.plot(x, np.sin(x), label ="Sin(x)");
plt.plot(x, np.cos(x), color = "red", label ="Cos(x")
plt.title("Sin(x) and Cos(x)");
plt.xlabel("Variable X");
plt.ylabel("Variable Y");
# plt.xlim([0,4])
# plt.ylim([1,2])
# plt.axis([0,4,-0.7, 0.6])
# plt.axis("tight")
plt.axis("equal")

plt.legend();
```

```python
# Object_Oriented API

#1. Prepare data
x =[1,2,3,4]
y =[11,22,33,44]

#2.Setup plot
fix, ax = plt.subplots(figsize=(5,5))

#3. Plot the data
ax.plot(x, y);
ax.set(title = "A simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis");
```

## Most common types of Matplotlib Plots
* Line
* Scatter
* bar
* subplots
* hist

# Line

```python
# Creating an array
x = np.linspace(0,10, 100)
x[:5]
```

```python
# OO API plot a line chart

fig, ax = plt.subplots()
ax.plot(x, x**3)
```

# Scatter

```python
# Pyplot API
plt.scatter(x, np.exp(x), color = "red")
```

```python
# OO API

# Prepare New Data
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s= 1000*rng.rand(100), c=rng.rand(100), alpha =0.6);

fig.colorbar(img1);
```

## 3.Bar
* Vertical
* Horizontal

```python
# Prepare Data from a python dictionary
soft_drink_prices = {"Coke": 10,
                     "Pepsi": 12,
                     "Sprite": 8}

fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax.set(title = "Drink Prices",
       ylabel = "Prices ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()));
```

# 4. Histogram

```python
#Prepare data
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
```

```python
student_height[:10]
```

```python
#pyplot API
plt.hist(student_height, bins=50);
```

# Subplots

```python
# Option 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))

#Plot data to each axis
ax1.plot(x, x/2);
ax1.plot(x, x+2);

ax2.scatter(np.random.random(10), np.random.random(10));
ax3.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()));
ax4.hist(student_height, bins=50);
```

```python
#option 2:
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize = (10,5))

ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x*2);

ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(list(soft_drink_prices.keys()), list(soft_drink_prices.values()));
ax[1,1].hist(student_height, bins=30);
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = pd.read_csv('/content/gdrive/MyDrive/california_cities.txt')
```

```python
cities.head()
```

```python
# Extract Latd and Longd

lat, lon = cities['latd'], cities['longd']
pop, area = cities['population_total'], cities['area_total_km2']

plt.style.use('seaborn');

#Plot using Pyplot API
plt.figure(figsize=(10,6));
plt.scatter(lon, lat, 
            c = np.log10(pop), cmap='viridis',
            s=area, linewidths=0, alpha = 0.7);
plt.axis('equal');
plt.xlabel('Longtitude');
plt.ylabel('Latitude');
plt.colorbar(label = 'log$_{10}$(pop)');
plt.clim(3,7);
plt.title("California Cities: Pupulation and Area Distribution");

#Create a legend for cities's size
area_range = [50,100,300,500]

for area in area_range:
  plt.scatter([], [], s=area,
              label = str(area)+ 'km$^2$',
              c='k', alpha = 0.6);
plt.legend(labelspacing =1, title ='Area scale');
```
